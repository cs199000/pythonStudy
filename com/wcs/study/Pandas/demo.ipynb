## 四、缺省值处理

```python
# reindex()方法可以对指定轴上的索引进行改变/增加/删除操作，这将返回原始数据的一个拷贝
df4 = df.reindex(index=pd.date_range('20170102', periods=3), 
                 columns=['B', 'D', 'E', 'F'])
df4.iloc[0: 2, 3] = 55
df4.iat[1, 2] = np.nan
df4
```

```python
# 删除值为NaN的行（axis=0）或列（axis=1）
df4.dropna(axis=0, how='any')  # how：any(有一个值为NaN就删)，all(全部)
```

```python
df4.fillna(0)
```

```python
df4.fillna(df4.mean())
```

```python
# 使用平均值填充NaN，可指定需要处理的列
df4.fillna(df4.mean()['F': 'F'])
```

```python
pd.isnull(df4)
```

## 五、相关操作

```python
df01 = pd.DataFrame([[1, '张三', 13], [2, '李四', 14], [3, '王五', 15]], columns=['id', 'name', 'age'])
df02 = pd.DataFrame([[1, 'math', 55], [1, 'English', 66], [2, 'math', 33], [4, 'math', 88]],
                    columns=['uid', 'subject', 'score'])
df03 = df01.merge(df02, how='right', left_on='id', right_on='uid')
df03
```

```python
df03.groupby('uid')['subject'].count()
```
