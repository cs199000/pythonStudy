# Matplotlib学习笔记

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# 折线图 plot
x = range(100)
plt.plot(x, [xi ** 2 for xi in x])
plt.show()
```

```python
x1 = np.arange(1000)
# plt.plot(x1, x1 ** 2, x1, (x1+100) * 600)
plt.plot(x1, x1 ** 2, label='x1 ** 2')
plt.plot(x1, (x1+100) * 600, label='(x1+100) * 600')
plt.plot(x1, x1 ** 2.1, label='x1 ** 2.1')
plt.legend()  # 图例
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Test')
plt.savefig('pic1.png')
plt.show()
```

```python
x2 = np.random.randint(50, 101, 12)
x3 = np.random.randint(50, 101, 12)
plt.plot(x2, label=u'张三')
plt.plot(x3, label=u'李四')
plt.xticks(range(0, 12), ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
                          'AUG', 'SEP', 'OCT', 'NOV', 'DEC'])
plt.yticks([60, 80, 90, 100], [u'及格', u'良', u'优', u'满分'])
plt.xlabel(u'月份')
plt.ylabel(u'成绩')
plt.title(u'张三、李四每月成绩图')
plt.legend()
plt.show()
```

```python
# 分布直方图 Histogram charts 使用hist方法
y = np.random.randn(1000)
plt.hist(y, bins=25)  # 分25组 bins
plt.show()
```

```python
# 误差图 Error bar charts
x4 = np.arange(0, 4, 0.2)
y4 = np.exp(-x4)
e1 = 0.1 * np.abs(np.random.randn(len(y4)))
e2 = 0.1 * np.abs(np.random.randn(len(y4)))
plt.errorbar(x4, y4, yerr=e1, xerr=e2)
plt.show()
```

```python
# Bar Chart 条线图
data1 = np.random.randint(10, 50, 8)
data2 = np.random.randint(10, 50, 8)
data3 = np.random.randint(10, 50, 8)

width = 0.25
locs = np.arange(1, len(data1)+1)
plt.figure(facecolor='w')
plt.bar(locs, data1, width=width, color='r', label='data1')
plt.bar(locs+width, data2, width=width, color='g', label='data2')
plt.bar(locs+width*2, data3, width=width, color='b', label='data3')
plt.xticks(locs+width, locs)
plt.xlabel(u'横坐标')
plt.ylabel(u'纵坐标')
plt.title(u'条图')
plt.grid(axis='both')
plt.legend()
plt.show()
```

```python
x6 = np.random.randint(10, 50, 5)
labels = ['A', 'B', 'C', 'D', 'E']
explode = [0.05, 0, 0, 0, 0.1]
plt.figure(figsize=(5, 5), facecolor='w')
plt.pie(x6, labels=labels, explode=explode, autopct='%1.1f%%')
plt.show()
```
