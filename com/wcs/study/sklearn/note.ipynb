```python
from sklearn import datasets
import pandas as pd
import os
```

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=['sepal-len', 'sepal-width', 'petal-len', 'petal-width'])
target = pd.DataFrame(iris.target)
data['target'] = target
data
```

```python
data.to_csv('iris.csv')
```
